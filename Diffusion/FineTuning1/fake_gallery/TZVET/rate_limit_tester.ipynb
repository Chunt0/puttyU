```python
import requests
import time

# Define the target URL and the number of requests to send
target_url = "https://tzvetnik.online/article/im-sorry-i-didnt-quite-catch-that"
num_requests = 1000  # You can adjust this number as needed

# Define the function to test the rate limit
def test_rate_limit(url, num_requests):
    headers = {
        "User-Agent": "Rate Limit Tester"  # Use a custom user-agent to identify your script
    }
    
    # Track response times for each request
    response_times = []

    for _ in range(num_requests):
        start_time = time.time()
        response = requests.get(url, headers=headers)
        end_time = time.time()

        # Calculate the response time for the request
        response_time = end_time - start_time
        response_times.append(response_time)

        # Print diagnostic information
        if response.status_code != 200:
            print(f"Request {len(response_times)} - Status Code: {response.status_code}, Response Time: {response_time:.2f} seconds")

        # Wait for a short interval before the next request (adjust as needed)
        #time.sleep(1)

    # Calculate and print some statistics
    avg_response_time = sum(response_times) / num_requests
    max_response_time = max(response_times)
    min_response_time = min(response_times)
    
    print("\nRate Limit Test Summary:")
    print(f"Average Response Time: {avg_response_time:.2f} seconds")
    print(f"Maximum Response Time: {max_response_time:.2f} seconds")
    print(f"Minimum Response Time: {min_response_time:.2f} seconds")

# Run the rate limit test
test_rate_limit(target_url, num_requests)
```
